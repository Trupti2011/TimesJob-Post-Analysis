```python
driver = webdriver.Chrome()

url = 'https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=&txtLocation=india'
driver.get(url)
```

```python
pages = np.arange(1,2)
n=[]
for page in pages:
    soup = BeautifulSoup(driver.page_source,'html5lib')
    result2 = result.find_all('li', class_='clearfix job-bx wht-shd-bx')
    for i in result2:
        # Title
        title=i.find('a')
        title=title.text
        print(title)   # .encode('utf-8')
        # Company
        company=i.find('h3',class_='joblist-comp-name')
        company=company.text
        print(company)   #.encode('utf-8')
        # Experience
        expe=i.find('i',class_='material-icons').next_sibling
        print(expe)
        # Salary
        salary=i.find('i',class_='material-icons').find_next().text.strip()
        #salary=salary.text
        print(salary)
        #location
        spans = i.find_all('span')
        location = spans[1].text
        
         #location=i.find_all('span')
            #location=location.text
        print(location)
        # Posted date
        dated=i.find('span',class_='sim-posted')
        dated=dated.text.strip()
        print(dated)
        #KeySkills
        skills=i.find('span',class_='srp-skills') #.strip()#,class_=' srp-skills')
        skill=skills.text.strip()
        print(skill)
       
        # description
        description=i.find('label').next_sibling.strip()
        print(description)
        # URL
        url=i.find('a').get('href')
        print(url)
        n.append([title,company,expe,salary,location,dated,skill,description,url])
   # driver.execute_script("window.scrollTo(0,(document.body.scrollHeight))")
   # driver.find_element(By.XPATH, '/html/body/div[3]/div[4]/section/div[2]/div[2]/div[4]/em[2]/a').click()
```
